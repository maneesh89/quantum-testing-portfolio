```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit import transpile
from qiskit.visualization import plot_histogram

simulator=AerSimulator()

test_cases = [
    ("|00⟩", [], "00"),           # No X gates → |00⟩ input
    ("|01⟩", [1], "01"),          # X on qubit 1 → |01⟩ input
    ("|10⟩", [0], "11"),          # X on qubit 0 → |10⟩ input
    ("|11⟩", [0, 1], "10")        # X on both → |11⟩ input
]

# testcases = [
# ("00", [], "00"),
# ("01", [0], "01"),
# ("10", [1], "11"),
# ("11", [0,1], "10")    
# ]

for state_input, x_gates, expected in testcases:
    # Create Circuit
    qc = QuantumCircuit(2, 2)

    # Prepare input state
    for qubit in x_gates:
        qc.x(qubit)
    
    #Apply C NOT Gate 
    qc.cx(0,1)

    qc.measure((0, 1), (1, 0))

    job = simulator.run(transpile(qc, simulator), shots = 1024)
    counts = job.result().get_counts()
    measured = max(counts, key = counts.get)

    if measured == expected:
        print("success")

    else:
        print("failed")
```

```python
# Test all four CNOT behaviors

test_cases = [
    ("|00⟩", [], "00"),           # No X gates → |00⟩ input
    ("|01⟩", [1], "01"),          # X on qubit 1 → |01⟩ input
    ("|10⟩", [0], "11"),          # X on qubit 0 → |10⟩ input
    ("|11⟩", [0, 1], "10")        # X on both → |11⟩ input
]

print("Testing CNOT on all four basis states:")
print()

for input_state, x_gates, expected_output in test_cases:
    # Create circuit
    qc = QuantumCircuit(2, 2)
    
    # Prepare input state
    for qubit in x_gates:
        qc.x(qubit)
    
    # Apply CNOT (control=0, target=1)
    qc.cx(0, 1)
    
    # Measure
    qc.measure([0, 1], [1, 0])

    
    # Run
    job = simulator.run(transpile(qc, simulator), shots=100)
    result = list(job.result().get_counts().keys())[0]
    
    status = "✅" if result == expected_output else "❌"
    print(f"{input_state} → CNOT → Expected: {expected_output}, Got: {result} {status}")

print()
print("CNOT confirmed working!")
```
