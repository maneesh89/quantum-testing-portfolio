Build the Circuit X then H and print circuit, reult and graph

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit import transpile
from qiskit.visualization import plot_histogram

#Build the CIrcuit

qc_xh= QuantumCircuit(1 ,1)

qc_xh.x(0)
print("X gate added")

qc_xh.h(0)
print("H gate added")
qc_xh.measure(0,0)

#Create Simulator
simulator=AerSimulator()

#Compile and Run

compile= transpile(qc_xh, simulator)
job=simulator.run(compile, shots=1024)
counts= job.result().get_counts()
print(f"Run: {counts}")

qc_xh.draw('mpl')


```

```python
plot_histogram(counts)
```

Calculate the Percentage

```python
count_0=counts.get('0', 0)
count_1=counts.get('1', 0)
print(f"'0':{count_0} times({count_0/1024*100: .1f} %)")
print(f"'1':{count_1} times({count_1/1024*100: .1f} %)")
```
