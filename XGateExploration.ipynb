```python
# Import what we need
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit import transpile
from qiskit.visualization import plot_histogram

print("✅ All tools imported!")
print("Ready to explore X gate")
```

# What is X Gate?
#
# X gate = quantum NOT gate = bit-flip gate
#
# What it does:
# - Flips |0⟩ to |1⟩
# - Flips |1⟩ to |0⟩
# - Like a light switch: OFF → ON, ON → OFF
#
# In vector notation:
# - [1, 0] → [0, 1]  (|0⟩ becomes |1⟩)
# - [0, 1] → [1, 0]  (|1⟩ becomes |0⟩)
#
# Matrix (from Week 2):
# X = [0  1]
#     [1  0]

# Comparing H and X Gates
#
# H GATE (Hadamard):
# - Creates superposition
# - [1, 0] → [0.707, 0.707]
# - Results: RANDOM (50/50)
# - Multiple runs give different results
#
# X GATE (NOT):
# - Creates definite state
# - [1, 0] → [0, 1]
# - Results: DETERMINISTIC (100% predictable)
# - Every run gives same result!
#
# Key insight: Not all quantum gates create randomness!

```python
# Visualize what X does
print("Starting state: |0⟩")
print("Vector form: [1, 0]")
print()
print("After X gate:")
print("State becomes: |1⟩")
print("Vector form: [0, 1]")
print()
print("What happened?")
print("→ Amplitude moved from |0⟩ slot to |1⟩ slot")
print("→ Complete flip!")
print("→ Like swapping the contents of two boxes")
```

```python
# Circuit 1: X gate + measurement
print("Building Circuit 1: X + Measure")
print()

# Step 1: Create circuit
qc_x = QuantumCircuit(1, 1)
print("✅ Circuit created: 1 qubit, 1 classical bit")

# Step 2: Apply X gate
qc_x.x(0)
print("✅ X gate applied to qubit 0")

# Step 3: Measure
qc_x.measure(0, 0)
print("✅ Measurement added")
```

```python
qc_x.draw('mpl')
```

# PREDICTION TIME!
#
# Question: What will we measure?
#
# Logic:
# 1. Qubit starts as |0⟩
# 2. X gate flips it to |1⟩
# 3. We measure it
#
# Prediction: We should get '1' every single time!
#
# Why? Because there's no superposition!
# The qubit is DEFINITELY in state |1⟩
#
# Let's test this prediction...

```python
# Run Circuit 1
print("Running X gate circuit...")
print("Shots: 1024")
print()

# Create simulator
simulator = AerSimulator()

# Compile and run
compiled_x = transpile(qc_x, simulator)
job_x = simulator.run(compiled_x, shots=1024)

# Get results
result_x = job_x.result()
counts_x = result_x.get_counts()

print("✅ Simulation complete!")
print()
print("Results:", counts_x)
```

```python
plot_histogram(counts_x)
```

```python
# Let's compare: Run H gate again
print("For comparison, let's run H gate circuit from yesterday:")
print()

# Create H circuit
qc_h = QuantumCircuit(1, 1)
qc_h.h(0)
qc_h.measure(0, 0)

# Run it
compiled_h = transpile(qc_h, simulator)
job_h = simulator.run(compiled_h, shots=1024)
counts_h = job_h.result().get_counts()

print("H gate results:", counts_h)
print("X gate results:", counts_x)
print()
print("Notice the difference?")
print("H gate: ~50/50 split (randomness)")
print("X gate: ~100% for '1' (certainty)")
```
