Apply X Gate 3 times and see the result

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit import transpile
from qiskit.visualization import plot_histogram

# Build the Circuit
qc_xxx= QuantumCircuit(1, 1)

qc_xxx.x(0)
qc_xxx.x(0)
qc_xxx.x(0)

qc_xxx.measure(0, 0)
qc_xxx.draw('mpl')

```

```python
#Create Simulator

simulator = AerSimulator()

#Compile and run
compile_xxx= transpile(qc_xxx, simulator)
job_xxx= simulator.run(compile_xxx, shot=1024)
counts_xxx=job_xxx.result().get_counts()

# Print result

print(f"Result: {counts_xxx}")
```

Experiment 2: H → X → H

```python
# Create Quantum Circuit H → X → H
qc_hxh= QuantumCircuit(1, 1)

# Apply H Gate
qc_hxh.h(0)

# Apply X Gate
qc_hxh.x(0)

# Apply H Gate
qc_hxh.h(0)

# Measure

qc_hxh.measure(0,0)

# Draw Circuit
qc_hxh.draw('mpl')

```

```python
# Create Simulator

simulator = AerSimulator()

# Compile and Run
compile_hxh = transpile(qc_hxh, simulator)
job_hxh= simulator.run(compile_hxh, shots = 1024)
counts_hxh = job_hxh.result().get_counts()

# Print

print(f"result: {counts_hxh}")
```
